```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# logistic regression
```

```python
class LogisticRegression: 
    def __init__(self, eta, epochs):
        self.eta = eta
        self.epochs = epochs
    
    def __init_weights__(self, n):
        self.weights = np.random.normal(loc= 0.1, scale = 0.1, size = n + 1)
        
    def net_weights(self, X):
        w_0 = self.weights[0]         
        nw = np.dot(X, self.weights[1:]) + w_0        
        return nw
    
    def activation_function(self, z):        
        sigmoid = 1.0 / (1.0 + np.exp(-z))       
        return sigmoid
    
    def error(self, y, y_):
        return y - y_
    
    def cost_function(self, output, y ):
        cost = np.dot(-y, np.log(output)) - np.dot((1-y), np.log(1- output))
        return cost
    
    def fit(self, X, y):
        n = X.shape[1]        
        self.__init_weights__(n)        
        costs = {}
        for i in range(self.epochs):
            nw = self.net_weights(X)
            z = self.activation_function(nw)
            e = self.error(y, z)
            self.weights[1:] += self.eta * np.dot(e, X)
            self.weights[0] += self.eta * e.sum()
            c = self.cost_function(z, y)
            costs[i] = c
        return costs
    
    def predict(self, X):
        nw = self.net_weights(X)        
        labels = np.where(nw > 0.0, 1, 0)
        return labels
        
        
```

```python
iris = pd.read_csv("iris.data", names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "label"])
```

```python
y = iris.iloc[0:100, 4].values
```

```python
y = np.where(y == "Iris-setosa",-1, 1)
X = iris.iloc[0:100, [0,2]].values
X.shape
```

```python
plt.scatter(X[:50, 0], X[:50, 1], marker = "o", color = "red", label = "setosa")
plt.scatter(X[50:, 0], X[50:, 1], marker = "x",  color = "blue", label = "versicolor")
plt.legend(loc = "upper left")
```

```python
l = LogisticRegression(0.001, 1000)
X.shape[1]
```

```python
costs = l.fit(X, y)
```

```python
c = costs.values()
```

```python
plt.plot(c)
```

```python
l.predict(X)
```
